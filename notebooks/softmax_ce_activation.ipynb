```python
import torch
import torch.nn as nn
import numpy as np



x = torch.tensor([2.0 , 3.0, 8.0], dtype=torch.float32)
output = torch.softmax(x, dim=0)
print(output)


# cross entropy

# remember cross entropy loss already applies softmax and nllloss (negative likelihood loss)
# So no softmax required in last layer

loss = nn.CrossEntropyLoss()

# True clas, it means that true calass is at first position i.e index 0
Y = torch.tensor([2, 0, 1])
print(f"True class : {Y}")

# n_samples x nclases = 3 x 3
Y_pred_good = torch.tensor([[0.1, 1.0, 2.1], [2.0, 1.0, 0.1], [0.1, 3.0, 0.1]])
Y_pred_bad = torch.tensor([[2.1, 1.0, 0.1], [0.1, 1.0, 2.1], [0.1, 3.0, 0.1]])

print(Y_pred_good.shape)

l1 = loss(Y_pred_good, Y)
l2 = loss(Y_pred_bad, Y)

print(f"entropy loss l1: {l1.item():.4f}")
print(f"entropy loss l2: {l2.item():.2f}")


prediction1 = torch.max(Y_pred_good, Y)
prediction2 = torch.max(Y_pred_bad, Y)

print(f"prediction 1: {prediction1}")
print(f"prediction 2: {prediction2}")

```
