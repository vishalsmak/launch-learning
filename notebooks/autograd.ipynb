```python
# pytorch tutorial 06 - Traning pipeline: Model, Loss, and Optimiser

import torch
import torch.nn as nn

# Step 0 : Training sample, watch the shape

X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)  # each row here is no of features and colums are no of samples . This is done considering 
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)  # inputs and outputs for model ...nn.Linear

n_samples, n_features = X.shape   # 4, 1

print(f"No of samples: {n_samples}, No of features: {n_features}")

X_test = torch.tensor([5], dtype=torch.float32)

input_size = n_features
output_size = n_features

learning_rate = 0.001
iterations = 3000

# Step 1: Design a model 

class LinearRegression(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegression, self).__init__()
        # define layers
        self.lin = nn.Linear(input_dim, output_dim)
        
model = LinearRegression(input_dim=input_size, output_dim=output_size)

print(f"Prediction Before training f(5) : {model.lin(X_test).item():.3f}")

# Step 2: Define loss and optimiser

loss = nn.MSELoss()   # mean square error loss from pytorch

optimiser = torch.optim.SGD(model.parameters(), lr = learning_rate) # Define an optimiser

# Step 3: Training loop

for epoch in range(iterations):
    
    y_predicted = model.lin.forward(X)
    
    lo = loss(Y, y_predicted)
    
    lo.backward()
    
    # update weights
    optimiser.step()
    
    # zero gradients
    optimiser.zero_grad()

    if epoch % 100 == 0:
        [w, b] = model.parameters()
        print(f"epoch = {epoch + 1},  w = {w[0][0].item():.3f}, loss = {lo:.8f}")
        print(w.item())
        
        
print(f"Prediction After training f(5) : {model.lin(X_test).item():.3f}")


```
